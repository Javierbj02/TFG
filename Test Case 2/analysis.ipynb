```python
def true_dist(x_pred, x_true):
    uno = x_pred[0] - x_true[0]
    uno = uno ** 2
    
    dos = x_pred[1] - x_true[1]
    dos = dos ** 2
    
    return (uno + dos) ** 0.5

# Cargar los datos
df_no_noise = pd.read_csv('C:/Users/javi2/Desktop/TFG - Webots/TFG/Test Case 2/Results/No ruido/ULA 8/predicciones_ULA8.csv')
df_low_noise = pd.read_csv('C:/Users/javi2/Desktop/TFG - Webots/TFG/Test Case 2/Results/Ruido bajo/ULA 8/predicciones_ULA8.csv')
df_mid_noise = pd.read_csv('C:/Users/javi2/Desktop/TFG - Webots/TFG/Test Case 2/Results/Ruido medio/ULA 8/predicciones_ULA8.csv')
df_high_noise = pd.read_csv('C:/Users/javi2/Desktop/TFG - Webots/TFG/Test Case 2/Results/Ruido elevado/ULA 8/predicciones_ULA8.csv')

# Calcular la distancia verdadera para cada dataset
df_no_noise['TrueDist'] = df_no_noise.apply(lambda x: true_dist([x['PredictedX'], x['PredictedY']], [x['RoundedX'], x['RoundedY']]), axis=1)
df_low_noise['TrueDist'] = df_low_noise.apply(lambda x: true_dist([x['PredictedX'], x['PredictedY']], [x['RoundedX'], x['RoundedY']]), axis=1)
df_mid_noise['TrueDist'] = df_mid_noise.apply(lambda x: true_dist([x['PredictedX'], x['PredictedY']], [x['RoundedX'], x['RoundedY']]), axis=1)
df_high_noise['TrueDist'] = df_high_noise.apply(lambda x: true_dist([x['PredictedX'], x['PredictedY']], [x['RoundedX'], x['RoundedY']]), axis=1)

# Calcular la media de las distancias verdaderas
mean_no_noise = df_no_noise['TrueDist'].mean()
mean_low_noise = df_low_noise['TrueDist'].mean()
mean_mid_noise = df_mid_noise['TrueDist'].mean()
mean_high_noise = df_high_noise['TrueDist'].mean()

print('Mean no noise:', mean_no_noise)
print('Mean low noise:', mean_low_noise)
print('Mean mid noise:', mean_mid_noise)
print('Mean high noise:', mean_high_noise)

# Preparar los datos para el boxplot
data = [df_no_noise['TrueDist'], df_low_noise['TrueDist'], df_mid_noise['TrueDist'], df_high_noise['TrueDist']]
labels = ['No noise', 'Low noise', 'Mid noise', 'High noise']
means = [mean_no_noise, mean_low_noise, mean_mid_noise, mean_high_noise]

# Crear el boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, palette="Set2")
plt.title('Mean error of ULA 8', fontsize=16)
plt.ylabel('Mean error [mm]', fontsize=14)
plt.xticks(ticks=[0, 1, 2, 3], labels=labels, fontsize=12)

# Aadir las medias al grfico
for i, mean in enumerate(means):
    plt.text(i, mean + 0.02, f'{mean:.2f}', horizontalalignment='center', size='medium', color='black', weight='semibold')

# Mostrar el grfico
plt.grid(True)
plt.show()
```
