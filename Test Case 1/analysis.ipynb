```python
descriptive_stats = data.describe()
print(descriptive_stats)
```

```python
# Desviación típica de cada columna
std_devs = data.std()
std_devs
```

```python
# Media de cada columna
means = data.mean()
means
```

```python
stat, p_value = shapiro(data.iloc[: ,0]) # Verificar normalidad para la primera columna
print(f"Shapiro-Wilk Test: Stat={stat}, p-value={p_value}")
```

```python
# Histograma de una columna (por ejemplo, la primera)
plt.hist(data.iloc[:, 0], bins=30, alpha=0.7)
plt.title('Distribución de Antena1_Portadora1_Modulo')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
correlation_matrix = data.corr()
correlation_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
